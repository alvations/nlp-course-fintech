```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed(1228)
%matplotlib inline
```

```python
num_topics = 50
```

```python
data = pd.read_pickle("../data/articles_lemmatized_no_SW.pkl")
```

```python
data = data[data.topic.isin(("Культура", "Общество", "Политика", "Экономика", "Спорт", "Мир", "Страна", "Армия", 
                            "Наука", "Гаджеты & Телеком", "Авто", "Москва"))] 


data_sample = pd.DataFrame(columns=data.columns)

for topic in data.topic.unique():
    sample = data[data.topic==topic].sample(100)
    data_sample = data_sample.append(sample)
    
print(len(data_sample))
```

```python
%%time
from gensim.corpora import *
texts = [data.body.iloc[i].split() for i in range(len(data))]
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
%%time
from gensim.models import  *
tfidf = TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
%%time
lsi = lsimodel.LsiModel(corpus=corpus_tfidf, id2word=dictionary, num_topics=num_topics)
```

```python
lsi.show_topics(15)
```

```python
%%time
lda = ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
```

```python
lda.show_topics(15)
```

```python
%%time
hdpmodel = HdpModel(corpus=corpus, id2word=dictionary)
```

```python
hdpmodel.show_topics()
```
